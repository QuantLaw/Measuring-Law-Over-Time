# Impact of consensus clustering on stability

```python
import pandas as pd
import matplotlib.pylab as plt
import seaborn as sns
```

```python
sns.set_style("whitegrid")
plt.rcParams['figure.figsize'] = (8,6)
plt.rcParams['font.size'] = 20
```

# US

```python
def make_boxplot(dataset, metric, ylabel, save_path=None):
    df = pd.read_pickle(f'../results/variance_impact_of_consensus_clustering_{dataset}.pickle')
    fig, ax = plt.subplots()
    ax.boxplot(
        [
            df.loc[metric,:].loc[i,:]['values'].tolist() 
            for i in sorted(df.loc[metric,:].index.unique())
        ], 
        notch=0,
        sym="", 
        medianprops=dict(color='r')
    )
    ax.set_xticklabels(sorted(df.loc[metric,:].index.unique()))
    plt.ylabel(ylabel)
    plt.xlabel("Runs for Consensus Clustering")
    plt.tight_layout()
    if save_path is not None:
        plt.savefig(save_path)
```

```python
make_boxplot('us_reg', 'NMI', 'Normalized Mutual Information',
             '../graphics/variance_impact_of_consensus_clustering_nmi_us_reg.pdf')
```

```python
make_boxplot('us_reg', 'Rand', 'Adjusted Rand Index', 
             '../graphics/variance_impact_of_consensus_clustering_rand_us_reg.pdf')
```
